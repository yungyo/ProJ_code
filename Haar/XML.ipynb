```python
import cv2
import os
import numpy as np
from PIL import Image

path = "D:/dbsry2021/rasp-record/back/camera_recorded_back" 
imagePaths = [os.path.join(path,file_name) for file_name in os.listdir(path)]
count = 0
sum = 0

for imagePath in imagePaths:
    count += 1
    filename = os.path.basename(imagePath)
    file_count = len(next(os.walk("D:/dbsry2021/rasp-record/front/camera_recorded_front/11"))[2])
    img = Image.open(imagePath)
    img_numpy = np.array(img, 'uint8')
    # 이미지 사이즈 수정 및 스케일링
    #img_scale = cv2.resize(img_numpy, dsize=(720,720), fx=1.3, fy=1.3, interpolation = cv2.INTER_CUBIC)
    gray = cv2.cvtColor(img_numpy, cv2.COLOR_BGR2GRAY)    
    
    xml = 'D:/boram/haarcascades/haarcascade_fullbody.xml'
    face_cascade = cv2.CascadeClassifier(xml)
    faces = face_cascade.detectMultiScale(gray, 1.01, 5)
    print(count,"/",file_count, "번째 이미지 " + "Number of faces detected: " + str(len(faces)))
   
    
    if len(faces):
        for (x,y,w,h) in faces:
            cv2.rectangle(img_numpy,(x,y),(x+w,y+h),(255,0,0),5)
            rgb = cv2.cvtColor(img_numpy, cv2.COLOR_BGR2RGB)
            cv2.imwrite("D:/boram/people/front_body/" + filename+ ".jpg" , rgb)
print("ALL DONE")
```
