```python
import os
import xml.etree.ElementTree as ET

#  

indir = 'D:\\boram\\people\\train\\annotations'
outdir = 'D:\\dbsry2021\\yolov5-master-test\\coco128\\labels\\501_'
num = 1

#indir .xml   
file_list = os.listdir(indir)
xml_list = [] # 
for file in file_list :
    if '.xml' in file :
        xml_list.append(file)

# .xml  
#xml_list  xml 
for xml_file in xml_list :
    target_path = indir + "/" + xml_file #  xml_file 
    targetXML = open(target_path, 'rt', encoding='UTF8')
    
    tree = ET.parse(targetXML) # (parse)
    root = tree.getroot()
    
    #  (iter)   
    for object in root.iter("object") : 
        target_tag = object.find("name")
        
        original = target_tag.text #
        target_tag.text = "person"  #text 
    
    print("[" + str(num) +"]" + xml_file + "[success]")
    
    end_path = outdir + "/" + xml_file
    tree.write(end_path)
    num += 1

print("finished")
```
