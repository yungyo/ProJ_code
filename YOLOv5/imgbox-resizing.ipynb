### 가로 길이 입력하면 그에 따라 세로 길이가 자동으로 바뀜

```python
import cv2
import os
import glob

files = glob.glob('D:/dbsry2021/capd_git/background_img/*.jpg')
#x_scale : 1 / y_scale : 0
for f in files:
    img = cv2.imread(f)
    #print(img.shape)
    want_garo = 640 #변경하고싶은 가로길이
    garo_b = want_garo/float(img.shape[1])
    #변경하고싶은 가로길이를 기존이미지 가로길이로 나누면 크기 비율 알 수 있다.
    #비율이 소수점이기에 float
    want_sero = int((float(img.shape[0])*float(garo_b)))
    #가로길이 비율을 기존 이미지 세로길이에 곱하면 변경하고싶은 세로길이를 구할 수 있다

    img_resize = cv2.resize(img,(want_garo,want_sero),interpolation = cv2.INTER_AREA)
    
    title, ext = os.path.splitext(f)
    cv2.imwrite(title+'_newb' + ext,img_resize)
    
print("finish")
```

# img resize + bounding box resize -> 최종

```python
import cv2
import os
import glob
import xml.etree.ElementTree as ET

files = glob.glob('D:/dbsry2021/capd_git/labels_xml/*.jpg')
xmls = glob.glob('D:/dbsry2021/capd_git/labels_xml/*.xml')


for f,xml in zip(files,xmls):
    img = cv2.imread(f)
    
    want_garo = 640 #변경하고싶은 가로길이
    garo_b = want_garo/float(img.shape[1])
    #변경하고싶은 가로길이를 기존이미지 가로길이로 나누면 크기 비율 알 수 있다.
    #비율이 소수점이기에 float
    want_sero = int((float(img.shape[0])*float(garo_b)))
    #가로길이 비율을 기존 이미지 세로길이에 곱하면 변경하고싶은 세로길이를 구할 수 있다

    img_resize = cv2.resize(img,(want_garo,want_sero),interpolation = cv2.INTER_AREA)
    title, ext = os.path.splitext(f)
    cv2.imwrite(title +'rs'+ ext,img_resize) 
    print(img_resize.shape)
    
    
    tree = ET.parse(xml)
    sizes = tree.findall('./size')
    for s, object_ in enumerate(sizes):
        width_tag = object_.find("width")
        original = width_tag.text #원본
        width_tag.text = str(want_garo) #text 변경
        
        height_tag = object_.find("height")
        original = height_tag.text #원본
        height_tag.text = str(want_sero) #text 변경
        
        objects = tree.findall('./object')
        for i, object__ in enumerate(objects):
            bndbox = object__.find('./bndbox')
            bndbox.find('./xmin').text =str(int(garo_b * float(bndbox.find('./xmin').text)))
            bndbox.find('./xmax').text =str(int(garo_b * float(bndbox.find('./xmax').text)))
            bndbox.find('./ymin').text =str(int(garo_b * float(bndbox.find('./ymin').text)))
            bndbox.find('./ymax').text =str(int(garo_b * float(bndbox.find('./ymax').text)))
        title, ext = os.path.splitext(xml)
        tree.write(title +'rs' + ext) #encoding='UTF8'
    
print("finish")
```
